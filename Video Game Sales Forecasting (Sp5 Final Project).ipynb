## 4.1 Regional Platform Analysis

We'll begin by examining platform performance across different regions:

```python
# Function to analyze platform performance by region

def platform_performance(dataframe, groupby, regions=['na_sales', 'eu_sales', 'jp_sales', 'other_sales'], operation='sum'):

    results = {}
    for region in regions:
        if operation == 'sum':
            results[region] = dataframe.groupby(groupby)[region].sum().reset_index()
        elif operation == 'mean':
            results[region] = dataframe.groupby(groupby)[region].mean().reset_index()
    return results   
```

```python
# Analyze each region
na_sales_platform_performance_avg = platform_performance(platforms_sales_2012_fwd, 'platform', ['na_sales'], 'mean')
na_sales_platform_performance_sum = platform_performance(platforms_sales_2012_fwd, 'platform', ['na_sales'], 'sum')
print('North America Average Sales by Platform:\n', na_sales_platform_performance_avg)
print()
print('North America Total Sales by Platform:\n', na_sales_platform_performance_sum)
print()


eu_sales_platform_performance_avg = platform_performance(platforms_sales_2012_fwd, 'platform', ['eu_sales'], 'mean')
eu_sales_platform_performance_sum = platform_performance(platforms_sales_2012_fwd, 'platform', ['eu_sales'], 'sum')
print('Europe Average Sales by Platform:\n', eu_sales_platform_performance_avg)
print()
print('Europe Total Sales by Platform:\n', eu_sales_platform_performance_sum)
print()

jp_sales_platform_performance_avg = platform_performance(platforms_sales_2012_fwd, 'platform', ['jp_sales'], 'mean')
jp_sales_platform_performance_sum = platform_performance(platforms_sales_2012_fwd, 'platform', ['jp_sales'], 'sum')
print('Japan Average Sales by Platform:\n', jp_sales_platform_performance_avg)
print()
print('Japan Total Sales by Platform:\n', jp_sales_platform_performance_sum)
print()

other_sales_platform_performance_avg = platform_performance(platforms_sales_2012_fwd, 'platform', ['other_sales'], 'mean')
other_sales_platform_performance_sum = platform_performance(platforms_sales_2012_fwd, 'platform', ['other_sales'], 'sum')
print('Other Regions Average Sales by Platform:\n', other_sales_platform_performance_avg)
print()
print('Other Regions Total Sales by Platform:\n', other_sales_platform_performance_sum)
print()
```

```python
# Function to analyze genre performance by region

def genre_performance(dataframe, groupby, regions=['na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'total_sales'], operation='sum'):

    results = {}
    for region in regions:
        if operation == 'sum':
            results[region] = dataframe.groupby(groupby)[region].sum().reset_index()
        elif operation == 'mean':
            results[region] = dataframe.groupby(groupby)[region].mean().reset_index()
    return results   
```
